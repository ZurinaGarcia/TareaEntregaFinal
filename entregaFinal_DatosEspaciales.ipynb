```python
import geopandas as gpd
import pandas as pd
import os
airports=gpd.read_file(ukraineMaps,layer='airports')
rivers=gpd.read_file(ukraineMaps,layer='rivers')
seaports=gpd.read_file(ukraineMaps,layer='seaports')
```

```python
airports
```

```python
# seleccionamos el subconjunto de los aeropuertos medianos
mediumAirports=airports[airports['type']=='medium_airport']
mediumAirports.reset_index(drop=True, inplace=True)

#ploteamos
base=mediumAirports.plot(color='red',marker="^")
seaports.plot(ax=base,alpha=0.5,markersize=3)
```

```python
seaports
```

```python
seaports.head()
```

```python
mediumAirports.head()
```

```python
distanceMatrixKM_sea_air= seaports.set_index('portName').geometry.apply\
                          (lambda g: mediumAirports.set_index('name').geometry.distance(g)/1000).\
                          sort_index(axis=0).sort_index(axis=1)
```

```python
distanceMatrixKM_sea_air.mean(axis=1).sort_values(ascending=True) 
```

```python
# puerto marÃ­timo mÃ¡s cercano a cada aeropuerto mediano
distanceMatrixKM_sea_air.idxmin(axis="rows")
```
